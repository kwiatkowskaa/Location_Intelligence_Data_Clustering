```python
import matplotlib.pyplot as plt

# Your category_colors dictionary
category_colors = {
    0: (0.267004, 0.004874, 0.329415, 1.0),
    1: (0.229739, 0.322361, 0.545706, 1.0),
    2: (0.127568, 0.566949, 0.550556, 1.0),
    3: (0.369214, 0.788888, 0.382914, 1.0),
    4: (0.993248, 0.906157, 0.143936, 1.0)
}

# Example value_counts, replace this with your actual value_counts
value_counts = {0: 10, 1: 20, 2: 15, 3: 25, 4: 18}

# Extracting counts and cluster numbers
counts = list(value_counts.values())
clusters = list(value_counts.keys())

# Creating the bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(clusters, counts, color=[category_colors[c] for c in clusters])

# Adding labels and title
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.title('Count of Instances in Each Cluster')

# Adding legend
legend_handles = [plt.Rectangle((0,0),1,1, color=category_colors[c]) for c in clusters]
plt.legend(legend_handles, ['Cluster {}'.format(c) for c in clusters])

# Show plot
plt.show()
```
